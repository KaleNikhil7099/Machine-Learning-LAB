```python
from datetime import date
date.today().year
data['age']=date.today().year-data['Year']
data.head()

```

```python
data['Brand']=data.Name.str.split().str.get(0)
data['Model']=data.Name.str.split().str.get(1)+data.Name.str.split().str.get(2)
data[['Name','Brand','Model']]
```

```python
print(data.Brand.unique())
```

```python
searchfor = ['Isuzu' ,'ISUZU','Mini','Land']
data[data.Brand.str.contains('|'.join(searchfor))].head(5)
data["Brand"].replace({"ISUZU": "Isuzu", "Mini": "Mini Cooper","Land":"Land Rover"}, inplace=True)
```

```python
plt.figure(figsize=(5, 3))
sns.histplot(data['Year'], kde=False, bins=20, color='pink', edgecolor='black')
plt.title('Car Count by Manufacturing Year')
plt.xlabel('Year')
plt.ylabel('Number of Cars')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
data.groupby('Brand')['Price'].mean().sort_values(ascending=False).head(10).plot(kind='bar')

plt.title('Top 10 Brands by Average Price')
plt.ylabel('Average Price (Lakhs)')
plt.xlabel('Brand')
plt.show()
```

```python


fuel_price = data.groupby('Fuel_Type')['Price'].mean().sort_values(ascending=False)

fuel_price.plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=140,
    figsize=(5 ,5)
)

plt.title('Average Price by Fuel Type')
plt.ylabel('')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Clean the 'Mileage' column
data['Mileage'] = data['Mileage'].str.extract(r'([\d.]+)').astype(float)

# Step 2: Group, average, sort, and select top 50
top_km = (
    data.groupby('Kilometers_Driven')['Mileage']
    .mean()
    .sort_values(ascending=False)
    .head(50)
    .sort_index()
)

# Step 3: Plot the data
plt.figure(figsize=(8, 4))
plt.scatter(top_km.index, top_km.values, color='red', s=15)
plt.title('Top 50 Kilometers Driven by Average Mileage', fontsize=12)
plt.xlabel('Kilometers Driven', fontsize=10)
plt.ylabel('Average Mileage (kmpl)', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
